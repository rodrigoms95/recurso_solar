```python
import os
import shutil
import pandas as pd
import warnings
warnings.filterwarnings("error")

temp_dir = "/home/rodr/temp/recurso_solar/duck_curve/folders/"
dest_dir = "/datos/rodr/Datos/NSRDB_2km_duck/"
#dest_dir = "/home/rodr/temp/recurso_solar/duck_curve/files/"

if not os.path.exists(temp_dir): os.mkdir(temp_dir)
if not os.path.exists(dest_dir): os.mkdir(dest_dir)

folders = os.listdir(temp_dir)
for folder_i in folders:
    print(folder_i)
    files = os.listdir(f"{temp_dir}{folder_i}")
    dir_n = [f"{dest_dir}{"_".join(x.split("_")[1:3])}" for x in files]
    files_n = [f"{"_".join(x.split("_")[1:])}" for x in files]
    for d in dir_n: 
        if not os.path.exists(d): os.mkdir(d)
    for i in range(len(files_n)):
        if not os.path.exists(files_n[i]):
            try: df = pd.read_csv(f"{temp_dir}{folder_i}/{files[i]}")
            except pd.errors.DtypeWarning: df = pd.read_csv(
                f"{temp_dir}{folder_i}/{files[i]}", skiprows = 2)
            finally: df.to_csv(f"{dir_n[i]}/{files_n[i]}", index = False)
    shutil.rmtree(f"{temp_dir}{folder_i}")
```
