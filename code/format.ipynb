```python
import os

import numpy as np
import pandas as pd

import xarray as xr
```

```python
# Cargamos los archivos.

lat = 19.45
lon = -99.14
years  = list( range(1998, 2023) )
months = list( range(1, 13) )

dir_name = f"../data/test/NSRDB/{lat}_{lon}/"
columns = [ "Year", "Month", "Day", "Hour", "Minute",
    "Temperature", "Dew Point", "GHI", "Wind Speed", "Wind Direction" ]

# Unimos todos los años.
df = pd.DataFrame( columns = columns )
for file in os.listdir(dir_name):
    df = df.append( pd.read_csv( dir_name + file, skiprows = 2 ) )

# Convertimos a fecha.
df["Time"] = pd.to_datetime( df["Year"].astype(str) + "/"
    + df["Month"].astype(str) + "/" + df["Day"].astype(str)
    + " " + df["Hour"].astype(str) + ":00:00" )

# Corregimos formato de columnas.
df = df.drop( columns[0:5], axis = 1 )

df["Latitud"] = lat
df["Longitud"] = lon

ds_1 = df.set_index( ["Time", "Latitud", "Longitud"] 
    ).astype(float).round( decimals = 1 ).to_xarray()

#ds.to_netcdf( f"../data/test/{lat}_{lon}.nc" )

ds_1
```
